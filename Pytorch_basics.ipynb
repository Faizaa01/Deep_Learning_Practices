```python
e = torch.randint(size=(3,2), low=0, high=10, dtype=torch.float32)
e
```

```python
print(torch.sum(e),
      torch.sum(e, dim=0),
      torch.sum(e, dim=1),
      torch.prod(e))
```

```python
print(torch.mean(e),
torch.mean(e, dim=1),
torch.mean(e, dim=0),
torch.median(e),
torch.std(e),
torch.var(e),
)
```

```python
print(torch.max(e),
  torch.min(e),
  torch.argmin(e), #index of min value
  torch.argmax(e), #index of max value
)
```
